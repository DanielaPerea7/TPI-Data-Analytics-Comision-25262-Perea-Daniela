### 5锔 Calidad de Datos

```python
#Funci贸n Control de Calidad de Datos


def calidad(df, nombre, clave=None):

    ## Mostrar t铆tulo descriptivo con el nombre del DF
    print(f"=== {nombre} ===")


    ## Mostrar cantidad de valores nulos por columna
    print("Valores nulos por columna:")
    display(df.isna().sum().to_frame("nulos"))

    ## Contar filas duplicadas completas
    dup_rows = df.duplicated(keep=False).sum()
    print("Filas duplicadas (exactas):", dup_rows)

    if clave and clave in df.columns:
        # Contar cu谩ntas filas tienen valores repetidos en esa columna
        dup_key = df[clave].duplicated(keep=False).sum()
        print(f"Duplicados por clave '{clave}':", dup_key)

        # Si existen duplicados, mostrar cu谩les son los valores m谩s repetidos
        if dup_key > 0:
            # Filtrar filas donde esa clave est茅 duplicado
            duplicados_ordenados = (
                df[df[clave].duplicated(keep=False)][clave]
                .value_counts()
                .sort_values(ascending=False)
            )

            print("\n Top valores duplicados m谩s frecuentes:")
            # Mostrar solo los primeros 10 (los m谩s repetidos)
            display(duplicados_ordenados.head(10))
        else:
            print(f"No se encontraron duplicados en la clave '{clave}'.")
    else:
        # Si la clave no fue pasada o no existe en el DataFrame
        if clave:
            print(f"La clave '{clave}' no existe en el DataFrame.")
        else:
            print("No se indic贸 una clave para analizar duplicados por columna.")
```

```python
## Evaluar calidad de datos del dataset de Ventas
calidad(ventas, "VENTAS", clave="id_venta")
```

```python
## Evaluar calidad de datos del dataset de Clientes
calidad(clientes, "CLIENTES", clave="id_cliente")
```

```python
## Evaluar calidad de datos del dataset de Marketing
calidad(marketing, "MARKETING", clave="id_campanha")
```

### 1锔 Limpieza de datos

```python
# Crear copias de dataframes
ventas_clean = ventas.copy()
clientes_clean = clientes.copy()
marketing_clean = marketing.copy()

# Eliminar filas completamente duplicadas
ventas_clean = ventas_clean.drop_duplicates()
clientes_clean = clientes_clean.drop_duplicates()
marketing_clean = marketing_clean.drop_duplicates()
```

```python
calidad(ventas_clean, "VENTAS CLEAN", clave="id_venta")
```

```python
# Funci贸n para limpiar texto en columnas tipo string
def normalizar_texto(df):
    for col in df.select_dtypes(include="object").columns:
        df[col] = (
            df[col]
            .astype(str)
            .str.strip()
            .str.replace(r"[\u200b\t\r\n]", "", regex=True)
            .str.replace(" +", " ", regex=True)
            .str.title()
        )
    return df
```

```python
# Normalizar fechas
for df in [ventas_clean, clientes_clean, marketing_clean]:
    for col in df.columns:
        if "fecha" in col.lower():
            df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
```

```python
# Normalizar fechas de df Marketing

marketing_clean["fecha_inicio"] = pd.to_datetime(marketing_clean["fecha_inicio"], errors="coerce", dayfirst=True)
marketing_clean["fecha_fin"] = pd.to_datetime(marketing_clean["fecha_fin"], errors="coerce", dayfirst=True)
```

```python
# Verificar el tipo de dato
print(ventas_clean.dtypes)
print()
print(clientes_clean.dtypes)
print()
print(marketing_clean.dtypes)
print()
```

```python
#  Aplicar la normalizaci贸n de texto
ventas_clean = normalizar_texto(ventas_clean)
clientes_clean = normalizar_texto(clientes_clean)
marketing_clean = normalizar_texto(marketing_clean)
```

```python
# Mostrar los df luego de normalizar los textos para revisar que queden bien
print(ventas_clean.head(10))
print(clientes_clean.head(10))
print(marketing_clean.head(10))
```

```python
# Normalizar valores num茅ricos

if "precio" in ventas_clean.columns:
    ventas_clean["precio"] = (
        ventas_clean["precio"]
        .astype(str)
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.strip()
    )
    ventas_clean["precio"] = pd.to_numeric(ventas_clean["precio"], errors="coerce")
```

```python
# Reporte Global de Calidad de Datos (de los 3 df)

def reporte_calidad_global(dfs, nombres):
    resumen = []
    for df, nombre in zip(dfs, nombres):
        nulos = df.isna().sum().sum()
        duplicados = df.duplicated(keep=False).sum()
        columnas = len(df.columns)
        filas = len(df)

        resumen.append({
            "Dataset": nombre,
            "Filas": filas,
            "Columnas": columnas,
            "Nulos totales": nulos,
            "Duplicados": duplicados,
        })

    reporte = pd.DataFrame(resumen)
    return reporte
```

```python
# Presentaci贸n del reporte

print(reporte_calidad_global([ventas, clientes, marketing], ["VENTAS Original", "CLIENTES Original", "MARKETING Original"]))
print(reporte_calidad_global([ventas_clean, clientes_clean, marketing_clean],["VENTAS Copia   ", "CLIENTES Copia   ", "MARKETING Copia   "]))
```

### 2锔 Transformaci贸n de datos

```python
# Funcion para encontrar columnas con nombres

def encontrar_columna(df, candidatos):
    for c in df.columns:
        nombre = c.lower()
        if any(p in nombre for p in candidatos):
            return c
    return None
```

```python
# Detectar la columna de producto

prod_col = encontrar_columna(ventas_clean, ["producto", "id_producto", "sku", "articulo", "art铆culo"])
if prod_col is None:
    raise ValueError("No se encontr贸 columna de producto. Renombr谩 una columna a 'producto' o similar.")
print(prod_col)
```

```python
# Calcular ingreso por registro = precio * cantidad

ventas_perf = (
    ventas_clean
    .assign(
        ingreso = ventas_clean["precio"] * ventas_clean["cantidad"]
    )
)
```

```python
# Agregar m茅tricas por producto

resumen_prod = (
    ventas_perf
    # 1) Agrupar el DataFrame por una o varias columnas clave
    .groupby(
        by=prod_col,
        dropna=False,
        as_index=False,
        observed=False
    )
    # 2) Agregamos (resumimos) columnas num茅ricas por cada grupo
    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        precio_promedio=('precio', 'mean'),
        registros=('ingreso', 'size')
    )
)
```

```python
# Verificar tabla de ventas

print(resumen_prod.head(10))
```

```python
# Calcular percentil 80 de ingreso_total

p80_ingreso = resumen_prod["ingreso_total"].quantile(
    q=0.80,
    interpolation="linear"
)
```

```python
# Filtrar los productos "de alto rendimiento" y ordenarlos

## Filtrar
ventas_top = (
    resumen_prod
    .query(
        "ingreso_total >= @p80_ingreso",
        engine="python"
    )

  ## Ordenar
    .sort_values(
        by=["ingreso_total", "unidades"],
        ascending=[False, False],
        na_position="last",
        ignore_index=True
    )
)

# Mostrar resultados
print(f"Columna de producto detectada: {prod_col}")
print(f"Umbral (percentil 80) de ingreso_total: {float(p80_ingreso):,.2f}")
print(" Productos de ALTO RENDIMIENTO (top 20% por ingreso):")
display(ventas_top.head(20))
```

### 3锔 Agregaci贸n

```python
# Funcion para encontrar la columna de categor铆a
def encontrar_columna(df, candidatos):

    for c in df.columns:
        if any(p in c.lower() for p in candidatos):
            return c
    return None
```

```python
# Detectar columna de categor铆a (acepta variantes)
cat_col = encontrar_columna(ventas_clean, ["categoria", "categor铆a", "categoria_producto", "rubro"])
if cat_col is None:
    raise ValueError("No se encontr贸 columna de categor铆a (por ej. 'categoria' o 'rubro').")
```

```python
# Asegurar columna 'ingreso' (si no existe, crearla)
if "ingreso" not in ventas_clean.columns:
    ventas_cat = ventas_clean.assign(ingreso = ventas_clean["precio"] * ventas_clean["cantidad"])
else:
    ventas_cat = ventas_clean.copy()
```

```python
# Agregaci贸n por categor铆a con groupby + agg
resumen_cat = (
    ventas_cat
    .groupby(
        by=cat_col,
        dropna=False,
        as_index=False
    )
    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        ventas=('ingreso', 'size'),
        precio_promedio=('precio', 'mean')
    )
    .sort_values(
        by='ingreso_total',
        ascending=False,
        na_position='last',
        ignore_index=True
    )
)
```

```python
# Ticket promedio por venta = ingreso_total / ventas
resumen_cat = resumen_cat.assign(
    ticket_promedio_por_venta = resumen_cat['ingreso_total'] / resumen_cat['ventas']
)

print("Columna de categor铆a detectada:", cat_col)
print("Resumen por categor铆a (ordenado por ingreso_total):")
display(resumen_cat.head(20))
```

###  *Insights principales*

- Ш **Electrodom茅sticos** lidera en **ingresos totales (505.000)**, combinando un volumen medio con el ticket promedio m谩s alto.  
-  **Decoraci贸n** presenta la **mayor cantidad de ventas (1,001)**, pero con el **ticket promedio m谩s bajo**, lo que sugiere compras m谩s frecuentes de bajo valor.  
-  **Electr贸nica** mantiene valores intermedios en todos los indicadores, mostrando un **equilibrio entre volumen y rentabilidad**.  
-  Los **precios promedios son similares (~$75.000)** entre categor铆as, lo que refleja una pol铆tica de precios coherente.  

###  *Conclusi贸n*

**Electrodom茅sticos** es la categor铆a **m谩s rentable** al combinar un **ticket promedio alto** con un **volumen s贸lido de ventas**, mientras que **Decoraci贸n** se destaca por la **frecuencia de compra** y **Electr贸nica** mantiene un **rendimiento estable y competitivo**.
